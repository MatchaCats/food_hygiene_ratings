## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
collections = db.list_collection_names()

# Pretty print the list of collections
pprint(collections)
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.
* Display the first document in the results using `pprint`.
* Convert the result to a Pandas DataFrame, print the number of rows in the DataFrame, and display the first 10 rows.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Query to find establishments with a hygiene score of 20 in the "scores" subdocument
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
first_document = establishments.find_one(query)
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Convert RatingValue from string to integer for all documents where it's currently a string
establishments.update_many(
    {"RatingValue": {"$type": "string"}},  # Only update documents where RatingValue is a string
    [{"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}]  # Convert RatingValue to integer
)

# Check a few documents to ensure the update was successful
sample_documents = establishments.find().limit(5)
for doc in sample_documents:
    pprint(doc)
```

```python
# Query to find establishments in London with RatingValue >= 4
query = {
    "LocalAuthorityName": "London",
    "RatingValue": {"$gte": 4}  # RatingValue should be >= 4
}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments in London with RatingValue >= 4: {count}")

# Display the first document in the results using pprint
first_document = establishments.find_one(query)
pprint(first_document)
```

```python
# Convert the result of the query to a Pandas DataFrame
df = pd.DataFrame(list(establishments.find(query)))  # Convert the cursor to a list and then to DataFrame

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of DataFrame:")
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Set the coordinates for the location of "Penang Flavours" (Latitude and Longitude)
latitude = 51.4769  # Example latitude for Greenwich
longitude = -0.0005  # Example longitude for Greenwich

# Set the degree of search range (0.01 degrees)
degree_search = 0.01

# Create the query to search within 0.01 degree of the given latitude and longitude
query = {
    "RatingValue": 5,  # RatingValue must be 5
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

# Sort by hygiene score (ascending order)
sort = [("scores.Hygiene", 1)]  # Sort by hygiene score from lowest to highest

# Limit the number of results to the top 5
limit = 5

# Execute the query and sort the results
results = establishments.find(query).sort(sort).limit(limit)

# Print the number of documents found using count_documents
num_results = establishments.count_documents(query)
print(f"Number of documents found: {num_results}")  # Show the count of documents found
```

```python
# Convert the query result to a Pandas DataFrame
df = pd.DataFrame(list(results))  # Use the actual variable 'results' here

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of DataFrame:")
print(df.head(10))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create the aggregation pipeline
pipeline = [
    {
        "$match": {  # Match documents with a hygiene score of 0
            "scores.Hygiene": 0
        }
    },
    {
        "$group": {  # Group by Local Authority and count the number of establishments
            "_id": "$LocalAuthorityName",  # Group by LocalAuthorityName
            "count": {"$sum": 1}  # Count the number of establishments in each Local Authority
        }
    },
    {
        "$sort": {  # Sort the results by the count of establishments in descending order
            "count": -1  # -1 for descending order
        }
    }
]

# Execute the aggregation query
results = establishments.aggregate(pipeline)

# Convert the aggregation results to a list
result_list = list(results)

# Print the number of documents in the result
print(f"Number of documents found: {len(result_list)}")

# Print the first 10 results
for result in result_list[:10]:
    pprint(result)
```

```python
# Convert the aggregation results to a Pandas DataFrame
df = pd.DataFrame(result_list)

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
df.head(10)
```
